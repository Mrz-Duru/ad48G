```r
#I merged my usstates datatable and usstates 2 for making a meaningful dcast and melt operation.
```

```r
usstates_big <- merge(usstates[, .(state.abb ,state.area , state.center.x , state.center.y)] , usstates2 , by.x = "state.abb" ,by.y = "state.short")
```

```r
#I have created a usstates_big datatable as a merge of usstates and usstates2 datatables.
```

```r
usstates_area <- usstates_big[, .(sa_max = max(state.area) , sa_min = min(state.area) , sa_av = mean(state.area)), by = c("state.division","state.region")]
usstates_area
```

```r
#I made a dcast operation I used state.division and state.region columns as edges and my value variable has become average value of state areas.
```

```r
states_area_wide <- dcast(usstates_area, state.division ~ state.region , value.var = "sa_av")
states_area_wide
```

```r
#I melted them properly.
```

```r
melt(states_area_wide , id.vars = "state.division" ,  variable.name = "state.region2" , value.name = "sa_av2")
```
