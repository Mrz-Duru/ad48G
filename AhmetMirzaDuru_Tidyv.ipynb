my second table for the purpose of using left join operation

```r
stackoverflow2 <- stackoverflow %>% select(Country , Remote , CareerSatisfaction)
```

filter operation

```r
stackoverflow %>% filter(Hobby == 1)
```

mutation operation

```r
stackoverflow %<>% mutate( salaryearsratio = Salary / Hobby )
```

summarise operation

```r
stackoverflow %>%  summarise( minratio = min(salaryearsratio))
```

group_by operation

```r
stackoverflow %>% group_by(Hobby , YearsCodedJob)  %>% summarise( minratio = min(salaryearsratio))
```

useful table for purpose of using pivot wider laterly.

```r
stat_salary <- stackoverflow %>%
group_by(Country, Salary) %>%
summarise(max_salary = max(Salary, na.rm = T),
          min_salary = min(Salary, na.rm = T),
          av_salary = mean(Salary, na.rm = T))
```

pivot wider operation

```r
stat_salary_wide <- stat_salary %>% pivot_wider(id_cols = Country, names_from = Salary, values_from = av_salary)
```

pivot wider output

```r
stat_salary_wide
```

pivot longer operation

```r
stat_salary_wide %>% pivot_longer(cols = -"Country", names_to = "Salary", values_to = "av_salary", values_drop_na = T)
```

left join operation and its output

```r
stackoverflow %<>% select(-Remote , -CareerSatisfaction) %>% left_join(stackoverflow2, by = "Country")
stackoverflow
```
